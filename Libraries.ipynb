```python
import numpy as np

arr = np.array([1, 2, 3, 4])
print(arr)
print(type(arr))
```

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([[1, 2], [3, 4]])

print(arr1.ndim)
print(arr2.ndim)
```

```python
print(arr1.shape)  # (3,)
print(arr2.shape)  # (2, 2)
```

```python
print(arr1.size)  # 3
print(arr2.size)  # 4
```

```python
#Matplotlib
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Line plot
plt.plot(x, y, label="Line")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic Line Plot")
plt.legend()
plt.show()
```

```python
#Scatter Plot
plt.scatter(x, y, color="red")
plt.title("Scatter Plot")
plt.show()
```

```python
#Bar Chart
categories = ["A", "B", "C", "D"]
values = [4, 7, 2, 5]

plt.bar(categories, values, color="green")
plt.title("Bar Chart")
plt.show()
```

```python
data = np.random.randn(1000)  # 1000 random values
plt.hist(data, bins=30, color="purple")
plt.title("Histogram")
plt.show()
```

```python
# Fewer bins (broad intervals)
plt.hist(data, bins=10, color="skyblue")
plt.title("10 Bins")
plt.show()

# More bins (narrow intervals)
plt.hist(data, bins=50, color="orange")
plt.title("50 Bins")
plt.show()
```
